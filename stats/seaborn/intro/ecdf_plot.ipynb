```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
tips.shape
```

## Basics

```python
sns.set_style('darkgrid')
```

```python
sns.ecdfplot(x = 'tip', data=tips );
plt.axvline(4, c='k')
```

```python
tips.tip.min(), tips.tip.max()
```

```python
tips.tip.value_counts().head()
```

```python
tips.tip.median()
```

```python
sns.ecdfplot(x = 'tip', data=tips, hue='time' );
```

```python
sns.ecdfplot(x = 'tip', data=tips, hue = 'day' );
```

```python
sns.ecdfplot(x = 'tip', data=tips, stat='count', hue='day' );
```

```python
sns.ecdfplot(x = 'tip', data=tips, hue='day', complementary=True)
```
