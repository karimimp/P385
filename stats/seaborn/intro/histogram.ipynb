```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = sns.load_dataset('penguins').dropna()
df.head(3)
```

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(df.flipper_length_mm)
```

```python
sns.histplot(y = 'flipper_length_mm', data=df)
```

```python
sns.histplot(x = 'flipper_length_mm', data=df, bins=20)
```

```python
sns.histplot(x = 'flipper_length_mm', data=df, bins=[170,200, 210,230])
```

```python
sns.histplot(x = 'flipper_length_mm', data=df,
            binwidth=10,
            binrange=(150, 240))
```

```python
sns.histplot(x = 'flipper_length_mm', data=df,
             stat='density'
            )
```

```python
sns.histplot(x = 'flipper_length_mm', data=df,
             stat='probability', fill=False, element='step'
            );
```

```python
sns.histplot(x = 'flipper_length_mm', data=df,
             stat='probability', fill=False,
             cumulative=True
             
            );
```

```python
sns.histplot(x = 'flipper_length_mm', data=df, hue='species', 
             element='step');
```

```python
sns.histplot(x = 'flipper_length_mm', data=df, hue='species', 
             element='poly');
```

```python
sns.histplot(x = 'flipper_length_mm', data=df, hue='species', 
             element='step',
            multiple='fill');
```

## Bivariate Histplot

```python
sns.histplot(x='flipper_length_mm', data=df,
             y='bill_length_mm', cbar=True, hue='species'
            );
```

## Styling

```python
sns.histplot(x = 'species', hue='sex', multiple='dodge', data=df,
             shrink=0.9, palette='bright'            
            );
```
