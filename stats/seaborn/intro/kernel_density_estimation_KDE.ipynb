```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head(3)
```

## Univariate (1D)

```python
sns.kdeplot(cars.horsepower)
```

```python
sns.kdeplot(cars.horsepower, shade=True)
```

```python
sns.kdeplot(cars.horsepower, shade=False, cumulative=True)
```

## Bivariate

```python
sns.kdeplot(cars.horsepower, y=cars.mpg, n_levels=10, shade=True);
```

```python
sns.kdeplot(cars.horsepower, y=cars.mpg, n_levels=10, shade=True, thresh=False);
```

```python
sns.kdeplot(cars.horsepower, y=cars.mpg, n_levels=10, 
            shade=True, 
            thresh=False,
           cbar=True);
```
