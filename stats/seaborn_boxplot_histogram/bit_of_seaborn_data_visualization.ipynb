```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.describe()
```

```python
sns.distplot(tips["total_bill"], kde=False).set_title("Hisogram of total plot")
```

```python
sns.distplot(tips["tip"], kde=True).set_title("Hisogram of tips")
```

```python
sns.distplot(tips["total_bill"], kde=False)
sns.distplot(tips["tip"], kde=False).set_title("Hisogram of total tips and bill")
```

```python
sns.boxenplot(data=tips["total_bill"]).set_title("Box plot of total bills")
```

```python
sns.boxenplot(data=tips["tip"]).set_title("Box plot of total tips")
```

```python
sns.boxenplot(x=tips["total_bill"]).set_title("Box plot of total bills")
sns.boxenplot(x=tips["tip"]).set_title("Box plot of total bills")
```

```python
sns.boxplot(x=tips["tip"], y=tips["smoker"])

g =  sns.FacetGrid(tips, row = "smoker")
g = g.map(plt.hist, "tip")
```

```python
sns.boxplot(x =  tips["tip"], y = tips["time"])

g = sns.FacetGrid(tips, row="time")
g = g.map(plt.hist, "tip")
```

```python
sns.boxplot(x = tips["tip"], y = tips["day"]).set_title("Tips vs Day of the Week")

g = sns.FacetGrid(tips, row="day")
g = g.map(plt.hist, "tip")
```

```python
sns.boxplot(x=tips["tip"], y=tips["sex"])

g = sns.FacetGrid(tips, row="sex")
g = g.map(plt.hist, "tip")
```
