```python
import pandas as pd
```

```python
data = pd.read_csv("NYC_Jobs.csv")
data.head(3)
```

```python
# apply either to the whole dataset or a sub-set of a dataset (1 column)
## change the 'Civil Service Title' from all capital to only first letter capital
def capitalize(title):
    return title.lower().capitalize()

data['Civil Service Title'].apply(capitalize)
```

```python
# applymap for whole dataframe and passes each element to the function
def add_year(text):
    return str(text)+'_2022'
data.applymap(add_year)
```
